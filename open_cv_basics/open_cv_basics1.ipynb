# OpenCV Basics: Edge Detection

This notebook explains three common edge detection methods in OpenCV:

- Sobel
- Laplacian
- Canny

Each method detects edges in a grayscale image, but they work differently and are useful for different tasks.

## 1. Setup

Import OpenCV and display images inline.

```python
import cv2
from matplotlib import pyplot as plt

def show_images(images, titles=None, cmap='gray'):
    n = len(images)
    plt.figure(figsize=(15, 5))
    for i, img in enumerate(images):
        plt.subplot(1, n, i + 1)
        plt.imshow(img, cmap=cmap)
        if titles:
            plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

## 2. Load a sample image

Use a grayscale image for edge detection. Replace `path_to_image` with a local path to your image file.

```python
image_path = 'path_to_image.jpg'  # change this to your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise FileNotFoundError(f'Image not found: {image_path}')

show_images([image], ['Original Grayscale Image'])
```

## 3. Sobel Edge Detection

Sobel uses first-order derivatives to compute gradients in the x and y directions. It is useful for detecting edges with orientation information.

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
sobel_combined = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

show_images([sobel_x, sobel_y, sobel_combined], ['Sobel X', 'Sobel Y', 'Sobel Combined'])
```

### How Sobel works

- `Sobel X` finds vertical edges.
- `Sobel Y` finds horizontal edges.
- Combining both gives a stronger edge map with orientation information.

## 4. Laplacian Edge Detection

Laplacian computes the second derivative of the image. It responds strongly to regions where intensity changes quickly, including both positive and negative transitions.

```python
laplacian = cv2.Laplacian(image, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

show_images([laplacian], ['Laplacian Edge Map'])
```

### How Laplacian works

- It uses the second derivative, so it is more sensitive to noise.
- It detects edges regardless of direction, but can produce stronger responses to small texture details.

## 5. Canny Edge Detection

Canny is a multi-stage detector that includes noise reduction, gradient calculation, non-maximum suppression, and edge tracking by hysteresis. It often produces clean edges.

```python
blurred = cv2.GaussianBlur(image, (5, 5), 1.4)
canny = cv2.Canny(blurred, threshold1=50, threshold2=150)

show_images([blurred, canny], ['Blurred Input', 'Canny Edges'])
```

### How Canny works

- `threshold1` and `threshold2` define the hysteresis thresholds.
- Strong gradients above the high threshold are kept as edges.
- Weak gradients between the thresholds are kept only if connected to strong edges.
- This yields clean and thin edge maps.

## 6. Summary

- Sobel: first derivative, directional edges.
- Laplacian: second derivative, direction-agnostic edge response.
- Canny: multi-step detector, produces clean, thin edges.
