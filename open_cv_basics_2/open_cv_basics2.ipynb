```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show_images(images, titles=None, cmap='gray', figsize=(16, 6)):
    n = len(images)
    plt.figure(figsize=figsize)
    for i, img in enumerate(images):
        plt.subplot(1, n, i + 1)
        if img is None:
            plt.text(0.5, 0.5, 'No image', ha='center', va='center')
            plt.axis('off')
            continue
        plt.imshow(img, cmap=cmap)
        if titles:
            plt.title(titles[i])
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

## 2. Load an Image for Thresholding

A grayscale image is ideal for thresholding. If you do not have a local image, the code creates a synthetic test image for demonstration.

```python
image_path = 'path_to_image.jpg'  # change this to a real image path if available
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
if image is None:
    # Create a synthetic grayscale image with text and shapes for thresholding
    image = np.zeros((320, 520), dtype=np.uint8)
    cv2.putText(image, 'OpenCV', (20, 120), cv2.FONT_HERSHEY_SIMPLEX, 3, 180, 6, cv2.LINE_AA)
    cv2.rectangle(image, (30, 180), (220, 300), 120, -1)
    cv2.circle(image, (400, 240), 60, 220, -1)
    cv2.line(image, (320, 30), (500, 120), 140, 10)

show_images([image], ['Input Grayscale Image'])
```

Thresholding converts a grayscale image into a binary image by mapping pixels below or above a threshold to fixed values.

OpenCV provides several threshold types: binary, inverse binary, truncation, to-zero, and to-zero inverse.

```python
_, th_binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
_, th_binary_inv = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
_, th_trunc = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)
_, th_tozero = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)
_, th_tozero_inv = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)

show_images([image, th_binary, th_binary_inv, th_trunc, th_tozero, th_tozero_inv],
            ['Original', 'Binary', 'Binary Inv', 'Trunc', 'To Zero', 'To Zero Inv'],
            cmap='gray', figsize=(18, 10))
```

### 3.1 What each threshold type does

- `THRESH_BINARY`: if pixel > T => 255, else 0.
- `THRESH_BINARY_INV`: if pixel > T => 0, else 255.
- `THRESH_TRUNC`: if pixel > T => T, else pixel.
- `THRESH_TOZERO`: if pixel > T => pixel, else 0.
- `THRESH_TOZERO_INV`: if pixel > T => 0, else pixel.

This gives you control over how the image is separated into background and foreground regions.

## 4. Adaptive Thresholding

For images with uneven illumination, a single global threshold may fail. Adaptive thresholding computes a local threshold for each pixel neighborhood.

OpenCV formulas for a pixel at `(x, y)` are:

- `THRESH_ADAPTIVE_THRESH_MEAN_C`: threshold = mean of neighborhood - C
- `THRESH_ADAPTIVE_THRESH_GAUSSIAN_C`: threshold = weighted Gaussian sum of neighborhood - C

This allows darker and brighter regions to be thresholded correctly in the same image.

```python
image_blur = cv2.GaussianBlur(image, (7, 7), 1.5)
th_adaptive_mean = cv2.adaptiveThreshold(image_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                         cv2.THRESH_BINARY, 15, 6)
th_adaptive_gauss = cv2.adaptiveThreshold(image_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                          cv2.THRESH_BINARY, 15, 6)

show_images([image, image_blur, th_adaptive_mean, th_adaptive_gauss],
            ['Original', 'Blurred', 'Adaptive Mean', 'Adaptive Gaussian'],
            cmap='gray', figsize=(16, 8))
```

## 5. Otsu's Thresholding

Otsu's method finds a threshold `T` that minimizes intra-class variance, or equivalently maximizes between-class variance.

If the histogram has two classes (background and foreground), Otsu chooses:

- `sigma_b^2(T) = w_0(T) w_1(T) [mu_0(T) - mu_1(T)]^2`

where `w_0` and `w_1` are class probabilities and `mu_0` and `mu_1` are class means.
